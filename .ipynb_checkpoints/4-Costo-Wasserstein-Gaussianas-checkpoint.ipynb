# Costo Wasserstein Gaussianas-

```python
%matplotlib inline
import numpy as np
import seaborn as sb
import matplotlib.pyplot as plt
from tpy.copulas import GaussianCopula

nsamples = 1000
l1 = 0
s1 = 2.0
r1 = 0.75

l2 = 1
s2 = 1.0
r2 = 0.95

gc1 = GaussianCopula(loc=l1, scale=s1, corr=r1, nsamples=nsamples)
gc2 = GaussianCopula(loc=l2, scale=s2, corr=r2, nsamples=nsamples)
```

```python
gc = gc1
```

```python
plt.figure(1, figsize=(16,5))

m1, m2 = gc.marginals()
J = gc.joint()
C = gc.copula()

plt.subplot(131)
_ = sb.distplot(m1)
_ = sb.distplot(m2)
plt.legend()

plt.subplot(132)
plt.plot(J[:,0], J[:,1], '*', label='Gaussian Distribution')
plt.legend()

plt.subplot(133)
plt.plot(C[:,0], C[:,1], '*', label='Gaussian Copula')
plt.legend()
```
