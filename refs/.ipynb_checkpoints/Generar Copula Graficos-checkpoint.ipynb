```python
%matplotlib inline
import g3py as g3
import numpy as np
import scipy as sp
import emcee
from scipy import stats as st
from tqdm import tqdm
import matplotlib.pyplot as plt
g3.style_normal()

dist1 = st.norm
dist2 = st.lognorm
dist2_param = 0.5

def clayton(u,v,theta=1):
    return ((theta+1)*(u*v)**(-(theta+1)))*(u**(-theta)+v**(-theta)-1)**(-(2*theta+1)/theta)

def copula(u,v, theta=1):
    c = np.empty(len(u))
    for i in range(len(u)):
        c[i] = clayton(u[i], v[i], theta)
    return c

def conjunta(u,v, theta=1):
    return copula(dist1.cdf(u), dist2.cdf(v, 1), theta)*dist1.pdf(u)*dist2.pdf(v, 1)
```

```python
samples = 1000
s1, s2 = dist1.rvs(size=samples), dist2.rvs(dist2_param, size=samples)
u1, u2 = dist1.cdf(s1), dist2.cdf(s2, dist2_param)

plt.figure(1, (15,15))

plt.subplot(331)
plt.scatter(s1,s2)

plt.subplot(332)
plt.scatter(u1,u2)

plt.subplot(333)
x = np.linspace(-4,4,1000)
plt.plot(x, dist1.cdf(x))

plt.subplot(334)
x = np.linspace(-1,7,1000)
plt.plot(x, dist2.cdf(x, dist2_param))

plt.subplot(335)
x = np.linspace(-4,4,1000)
plt.plot(x, dist1.pdf(x))

plt.subplot(336)
x = np.linspace(-1,7,1000)
plt.plot(x, dist2.pdf(x, dist2_param))
```

```python
ntemps = None
chains = 30
ndim = 2
nsamples = 5000
thin = 50
burnin = nsamples//2
start = np.array([0.49, 0.51])

theta = 10

def logp_copula_indep(x):
    if np.min(x)<0 or np.max(x) > 1:
        return -1e32
    return 0

def logp_copula(x):
    #print(clayton(x[0], x[1]), x)
    if np.min(x)<0 or np.max(x) > 1:
        return -1e32
    return np.log(clayton(x[0], x[1], theta))

def logp_independence(x):
    r = np.log(dist1.pdf(x[0])) + np.log(dist2.pdf(x[1], dist2_param))
    if np.isfinite(r):
        return r
    else:
        return -1e32

def logp_conjunta(x):
    r = np.log(clayton(dist1.cdf(x[0]), dist2.cdf(x[1], dist2_param), theta)) + np.log(dist1.pdf(x[0])) + np.log(dist2.pdf(x[1], dist2_param))
    if np.isfinite(r):
        return r
    else:
        return -1e32
    
    
def plot_samples(logp=logp_copula):
    print('Sampling {} variables, {} chains, {} times ({} temps)'.format(ndim, chains,nsamples,ntemps))
    noise = np.random.normal(loc=1, scale=0.1, size=(chains, ndim))
    p0 = noise * np.ones((chains, 1)) * start
    p0 += (p0 == 0)*np.random.normal(loc=0, scale=0.01, size=p0.shape)
    sampler = emcee.EnsembleSampler(chains, ndim, logp)
    for result in tqdm(sampler.sample(p0, iterations=nsamples), total=nsamples):
        pass
    lnprob, echain = sampler.lnprobability, sampler.chain
    C = echain[:,burnin::thin,:].reshape(-1,2)
    return C
    c1, c2 = C[:,0], C[:,1]
    plt.figure(1, (5,5))
    plt.scatter(c1,c2)
    plt.figure(2, (5,5))
    _ = plt.hist(c1, bins=100, normed=True, stacked=True)
    plt.figure(3, (5,5))
    _ = plt.hist(c2, bins=100, normed=True, stacked=True)
    
def plot_C(C):
    c1, c2 = C[:,0], C[:,1]
    plt.figure(1, (5,5))
    plt.scatter(c1,c2)
    plt.figure(2, (5,5))
    _ = plt.hist(c1, bins=100, normed=True, stacked=True)
    plt.figure(3, (5,5))
    _ = plt.hist(c2, bins=100, normed=True, stacked=True)
```

```python
C_copula_indep = plot_samples(logp_copula_indep)
```

```python
C_independence = plot_samples(logp_independence)
```

```python
C_conjunta = plot_samples(logp_conjunta)
```

```python
C_copula = plot_samples(logp_copula)
```

```python
plot_C(C_independence)
```

```python
plot_C(C_copula)
```

```python
plot_C(C_conjunta)
```

```python
s1, s2 = C_independence[:,0], C_independence[:,1]
u1, u2 = C_copula_indep[:,0], C_copula_indep[:,1]

plt.figure(1, (15,15))

plt.subplot(335)
plt.scatter(s2, s1)
plt.subplot(337)
plt.scatter(u2, u1)


plt.subplot(338)
x = np.linspace(-1,7,1000)
plt.plot(x, dist2.cdf(x, dist2_param))
plt.subplot(334)
x = np.linspace(-4,4,1000)
plt.plot(dist1.cdf(x), x)



plt.subplot(332)
x = np.linspace(-1,7,1000)
plt.fill(x, dist2.pdf(x, dist2_param))
plt.subplot(336)
x = np.linspace(-4,4,1000)
plt.fill(dist1.pdf(x), x)

plt.suptitle('Distribution with Independence Copula', size=30, position=(0.5,0.95)) 
g3.plot_save('/grios-sync/Dropbox/Latex/Thesis PhD/Examen Calificacion/images/copula_demo1.pdf')
```
